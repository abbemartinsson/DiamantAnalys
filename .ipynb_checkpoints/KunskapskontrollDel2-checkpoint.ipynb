# Inledning
Guldfynd vill utöka sitt produktsortiment till att inkludera diamanter. För att göra detta på ett säkert sätt krävs en analys av diamantmarknaden. Denna rapport syftar till att undersöka olika faktorer som påverkar diamantpriser, såsom vikt, färg, klarhet och slipning med målet att identifiera mönster och insikter som kan hjälpa Guldfynd i sin investering.

# Dataförberedelse och Datarensning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import streamlit as st

df = pd.read_csv("diamonds.csv")
 
df.dropna(how="any", inplace=True) # Tar bort nullvärden
 
df = df[(df['x'] > 0) & (df['y'] > 0) & (df['z'] > 0)] # Ta bort nollor eller negativa värden

df["depth_calc"] = 2 * (df["z"] / (df["x"] + df["y"])) * 100 #Beräknar depth med en formel
df["depth_diff"] = np.abs(df["depth"] - df["depth_calc"]) #Jämför den beräknande depthen mot den depthen som faktiskt finns
felaktiga_depthvärden = df[df["depth_diff"] > 1]#Visar depthvärden som skiljer sig med 1 procent eller mer
df = df.drop(felaktiga_depthvärden.index, errors='ignore') #Ta bort de depthvärdena som skiljer sig med 1 procent eller mer
```

# Caratplots

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# antal diamant i carat
bins = np.arange(0.5, df['carat'].max() + 0.5, 0.5)
sns.histplot(df['carat'], bins=bins, kde=False, ax=axes[0])
axes[0].set_title('Fördelning av carat')
axes[0].set_xlabel('carat')
axes[0].set_ylabel('Antal')
axes[0].set_xticks(bins)

# carat vs pris
sns.scatterplot(x='carat', y='price', data=df, ax=axes[1])
axes[1].set_title('Samband mellan carat och pris')
axes[1].set_xlabel('carat')
axes[1].set_ylabel('Pris')

plt.tight_layout()
plt.show()
```

### Vad ser man?
I den vänstra plotten ser att majoriteten av diamanterna har en vikt 0.5 och 1 carat. Antalet diamanter som har en vikt över 1 carat är väldigt få jämfört med de under 1 carat. Detta betyder att stora diamanter är mycket svårare att hitta på marknaden. I den högra plotten ser vi att det finns ett tydligt samband mellan carat och pris. Ju större carat en diamant har, desto dyrare är den. För diamanter med vikt under 1 carat är priset väldigt lågt jämfört med diamanter som är över 1 carat. För de större diamanterna som har en vikt på över 1 carat ser vi att priserna stiger snabbt.

### Varför ser det ut såhär?
Det ser ut såhär för större diamanter är mer sällsynta och får då ett mycket högre pris. Den vänstra plotten visar att mindre diamanter är vanligare vilket är logiskt eftersom de är lättare att hitta och för att de är billigare så fler personer har råd att köpa de.

### Vad kan vi göra med denna informationen?
De vi kan göra med denna informationen är att skapa en bra inköpsstrategi. Det kan vara mer lönsamt att fokusera på diamanter mellan 1 och 2 carat då de är relativt vanliga, antagligen har de en hyfsat hög efterfrågan och de finns i en bred pris separation.

# Korrelationsmatris

```python
# Filtrera för carat mellan 1 och 2
df_filtered = df[(df["carat"] >= 1) & (df["carat"] <= 2)]

# Skapa kopia
df_corr = df_filtered.copy()

df_corr["color"] = pd.Categorical(df_corr["color"], categories = ["D", "E", "F", "G", "H", "I", "J"], ordered=True)
df_corr["clarity"] = pd.Categorical(df_corr["clarity"], categories=["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"], ordered=True)
df_corr["cut"] = pd.Categorical(df_corr["cut"], categories=["Fair", "Good", "Very Good", "Premium", "Ideal"], ordered=True)

# Omvandla kategorier till numeriska värden
df_corr["color_num"] = df_corr["color"].cat.codes
df_corr["clarity_num"] = df_corr["clarity"].cat.codes
df_corr["cut_num"] = df_corr["cut"].cat.codes

corr = df_corr[["carat", "depth", "price", "color_num", "clarity_num", "cut_num"]].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korrelationsmatris mellan 1 och 2 carat')
plt.show()
```

### Vad ser man?
I denna korrelationsmatisen ser vi att carat och pris har den starkaste korrelationen på 0.92, vilket bekräftar de vi såg i scatterplotten med sambandet mellan carat och pris, alltså att större diamanter är mycket dyrare. Den näst starkaste korrelationen med pris är klarhet. 
Vi ser även att de finns väldigt svaga korrelationer till pris till exempel depth och färg.

### Vad kan vi göra med denna informationen?
Med denna informationen kan vi skapa en förbättrad inköpsstrategi där vi fokuserar på både caraten och klarheten på en diamant. 

# Klarhetsplots

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pris per klarhet
sns.boxplot(
    x="clarity",
    y="price",
    data=df_filtered,
    order=["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"],
    ax=axes[0]
)
axes[0].set_title("Pris per klarhet i 1–2 carat")
axes[0].set_ylabel("Pris")
axes[0].set_xlabel("Klarhet")
axes[0].grid(True)

# Antal per klarhet
clarity_counts = df_filtered["clarity"].value_counts().reindex(
    ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
)
sns.barplot(
    x=clarity_counts.index,
    y=clarity_counts.values,
    ax=axes[1]
)
axes[1].set_title("Antal diamanter per klarhet i 1–2 carat")
axes[1].set_ylabel("Antal")
axes[1].set_xlabel("Klarhet")
axes[1].grid(True)

# Layout och visa
plt.tight_layout()
plt.show()
```

## Vad ser man?
Vi ser att priset ökar från lägre till högre klarhet. Prisspridningen är bred inom de olika klarhetsnivåerna, speciellt i SI2, SI1, VS2 och VS1. Det finns prisintervall som korsar varandra mellan flera klarhetsnivåer, till exempel så kan vissa diamanter med klarhetsnivå "SI2" vara dyrare än de med högre klarhet. Detta kan också bero på andra faktorer, till exempel vikten. Som vi såg i korrelationsmatrisen så är caraten den viktigaste faktorn på priset av en diamant. Vi ser även att de tre vanligaste klarhetsnivåerna är SI2, SI1 och VS2. 

## Vad kan vi göra med denna informationen?
Med denna informationen kan vi forsätta på vår inköpsstrategi där som sagt fokuserar på carat och klarhet. Enligt plottsen kan vi komma fram till att de är bäst att fokusera på diamanter med vikt mellan 1-2 carat och klarhetsnivå mellan SI1 och VS2. Eftersom SI1 och VS2 är väldigt vanliga och inte är jätte dyra men fortfarande har en väldigt bra klarhet så finns det mer vinstmarginal att köpa in dessa klarhetsnivåerna.

# Slipningsplots

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot: Pris per slipning
sns.boxplot(data=df_filtered, x="cut", y="price", ax=axes[0])
axes[0].set_title("Pris per Slipning i 1-2 carat")
axes[0].set_xlabel("Slipning")
axes[0].set_ylabel("Pris")
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True)

# Piechart: Fördelning av slipning
cut_counts = df_filtered["cut"].value_counts()
axes[1].pie(
    cut_counts,
    labels=cut_counts.index,
    autopct='%1.1f%%',
    startangle=140
)
axes[1].set_title("Fördelning av Slipning i 1-2 carat")

plt.tight_layout()
plt.show()
```

## Vad ser vi?
Vi ser att Ideal, Premium och Very Good slipningar har högre priser. Det finns många outliers vilket betyder att att vissa diamanter har ett betydligt högre pris än genomsnittet för sin slipning. Men som vi vet från tidigare plots så finns det andra faktorer som är starkare än slipningen när det kommer till priset på en diamant. Vi ser även att de vanligaste slipningen är Ideal och Premium följt av Very Good. 

## Varför är det såhär?
Det ser ut såhär för Premium, Ideal och Very Good Ser bättre ut och ger därför diamanten ett högre pris. Att Ideal och Premium är den vanligaste slipningen kan bero på att tillverkare fokuserar på dessa typer av slipning då de antagligen är mest eftertraktade i marknaden.

## Vad kan vi göra med denna informationen?
Med denna informationen vet vi nu att vi ska fokusera på Ideal och Premium slipningar då de är vanligast och mest eftertraktade i marknaden.

# Lardomar och sammanfattning

Det vi har lärt oss med denna analys är att carat har den starkaste påverkan på priset. Vikten på en diamant har en väldigt stor påverkan på priset. Små skillnader i vikt kan leda till stora prisskillnader. Vi har lärt oss att det är bäst att köpa in diamanter med 1-2 carat. Vi har också lärt oss att klarhet har en hyfsat stor påverkan på priset. Inte lika stor som vikten men fortfarande rätt stor påverkan. Klarhetsnivån som rekommenderas att köpa in är SI1 och VS2 eftersom de är vanliga och har stor vinstmarginal. Vi vet nu också att diamantens slipning påverkar priset men inte alls lika mycket som klarheten eller vikten på en diamant. Det vi har kommit fram till är att om man ska fokusera på slipningen av en diamant vill man köpa in de som har en slipning på ideal och premium eftersom de är det vanligaste och mest eftertraktade i marknaden.
